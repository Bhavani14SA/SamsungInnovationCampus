```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

data = pd.read_csv('restaurants.csv')
print(data.head())


avg_price = data['Price'].mean()
print(f"Average Meal Price: ₹{avg_price:.2f}")

# Step 3: Correlation between Food and Service Quality
correlation = data[['Food_Quality', 'Service_Quality']].corr()
print("\nCorrelation Matrix:")
print(correlation)

# Scatter plot
plt.scatter(data['Food_Quality'], data['Service_Quality'], color='teal')
plt.title('Food Quality vs Service Quality')
plt.xlabel('Food Quality')
plt.ylabel('Service Quality')
plt.grid(True)
plt.show()

# Step 4: Regression Model
X = data[['Food_Quality', 'Service_Quality']]
y = data['Price']

model = LinearRegression()
model.fit(X, y)
predictions = model.predict(X)

# Model coefficients
print("\nRegression Coefficients:")
print(f"Intercept: {model.intercept_:.2f}")
print(f"Food Quality Coefficient: {model.coef_[0]:.2f}")
print(f"Service Quality Coefficient: {model.coef_[1]:.2f}")

# Model performance
mse = mean_squared_error(y, predictions)
print(f"\nMean Squared Error: {mse:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('garbage.csv')
print(data.head())

# Set plot style
sns.set(style="whitegrid")

# Plot histograms for numerical features
numerical_cols = ['weight', 'volume', 'moisture_content', 'organic_content']
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(data[col], kde=True, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

# Plot count plots for categorical features
categorical_cols = ['source', 'waste_type', 'label']
for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=data[col], palette='Set2')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset
data = pd.read_csv('restaurants.csv')
print(data.head())


# Set style
sns.set(style="whitegrid")

# Plot histograms
for col in ['Food_Quality', 'Service_Quality', 'Price']:
    plt.figure(figsize=(6, 4))
    sns.histplot(data[col], kde=True, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()

# Plot boxplots to detect outliers
for col in ['Food_Quality', 'Service_Quality', 'Price']:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=data[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```
